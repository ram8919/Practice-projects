```python
data_columns=['Country','Region','Happiness Rank','Happiness Score','Economy (GDP per Capita)','Family','Health (Life Expectancy)','Freedom','Trust (Government Corruption)','Generosity','Dystopia Residual']
```

```python
data=data[data_columns].copy()
```

```python
happy_data=data.rename({'Country':'Country_name','Region':'Regional','Happiness Rank':'Happiness_Rank','Happiness Score':'Happiness_Score','Economy (GDP per Capita)':'GDP_per_Capita','Family':'Family_Rank','Health (Life Expectancy)':'Life_Expectancy','Freedom':'Freedom_life','Trust (Government Corruption)':'Government_Corruption','Generosity':'Generosity','Dystopia Residual':'Dystopia_Residual'},axis=1)
```

```python
plt.rcParams['figure.figsize']=(15,10)
plt.title('plot between Happiness Score and GDP_per_Capita')
sns.scatterplot(x = happy_data.Happiness_Score ,y = happy_data.GDP_per_Capita,hue =happy_data.Regional,s=200);

plt.legend(loc='upper left',fontsize='12')
plt.xlabel('Happiness Score')
plt.ylabel('Economy(GDP per Capita)')
plt.rcParams['figure.facecolor']='blue'
sns.set_style('darkgrid')
```

```python
sns.scatterplot(x='Region',y='Happiness Rank',data=data)
plt.rcParams['figure.figsize']=(10,15)
plt.title('plot between Happiness Score and GDP_per_Capita')

plt.xlabel('Happiness Score')
plt.ylabel('Economy(GDP per Capita)')
plt.rcParams['figure.facecolor']='blue'
sns.set_style('darkgrid')
```

```python
GDP_region=happy_data.groupby('Regional')['GDP_per_Capita'].sum()
GDP_region
```

```python
GDP_region.plot.pie(autopct='%1.1f%%')
plt.title('GDP by region')
plt.ylabel('')
```

```python
Corruption=happy_data.groupby('Regional')[['Government_Corruption']].mean()
Corruption    
```

```python
plt.rcParams['figure.figsize']=(12,8)
plt.title('Perception of corruption in verious Regions')
plt.xlabel('Regions', fontsize=15)
plt.ylabel('Corruption Index',fontsize=15)
plt.xticks(rotation=30,ha='right')
plt.bar(Corruption.index, Corruption.Government_Corruption)
```

```python
plt.rcParams['figure.figsize']=(15,10)
sns.scatterplot(x = happy_data.Freedom_life ,y = happy_data.Happiness_Score,hue =happy_data.Regional,s=200);

plt.legend(loc='upper left',fontsize='12')
plt.xlabel('freedom of life')
plt.ylabel('happiness_score')
plt.rcParams['figure.facecolor']='blue'
sns.set_style('darkgrid')
```

```python
Country=happy_data.sort_values(by='Government_Corruption').head(10)
plt.rcParams['figure.figsize']=(12,8)
plt.title('Countries with most Government_Corruption')
plt.xlabel('Regions', fontsize=15)
plt.ylabel('Corruption Index',fontsize=15)
plt.xticks(rotation=30,ha='right')
plt.bar(Country.Country_name, Country.Government_Corruption)
```

```python
Country=happy_data.sort_values(by='Government_Corruption').tail(10)
plt.rcParams['figure.figsize']=(12,8)
plt.title('Countries with most Government_Corruption')
plt.xlabel('Regions', fontsize=15)
plt.ylabel('Corruption Index',fontsize=15)
plt.xticks(rotation=30,ha='right')
plt.bar(Country.Country_name, Country.Government_Corruption)
```
